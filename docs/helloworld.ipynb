```python
%matplotlib inline

from voromesh import Voronoi
from numpy import array

points = array([[-0.5, -0.5], [0.5, -0.5],
                [-0.5, 0.5], [0.5, 0.5],
                [0.0, 0.0]])

voro = Voronoi(points, buffer_size=0.4)

voro.plot()
```

```python
import pyvista as pv

mesh = voro.to_vtk()

# Convert mesh to PyVista UnstructuredGrid
grid = pv.UnstructuredGrid(mesh)

# Plot the grid in notebook
pv.set_jupyter_backend('static')
grid.plot(notebook=True, show_edges=True)
```
